```python
import matplotlib.pyplot as plt
import numpy as np

def draw_mlp():
    fig, ax = plt.subplots(figsize=(10, 6))

    # Define the number of neurons in each layer
    input_neurons = 3
    hidden_neurons_1 = 5
    hidden_layers = 1
    hidden_neurons_2 = 5
    output_neurons = 3

    # Define the layers
    layers = [input_neurons, hidden_neurons_1] + [hidden_layers] + [hidden_neurons_2, output_neurons]

    # Define layer names
    layer_names = ['Input Layer', '1st Hidden Layer', '...', 'Last Hidden Layer', 'Output Layer']

    # Draw the neurons and layers
    for i, layer in enumerate(layers):
        if layer == hidden_layers:
            ax.text(i, 0.5, '...', ha='center', va='center', fontsize=24)
            continue
        y_positions = np.linspace(0, 1, layer)
        for y in y_positions:
            ax.scatter(i, y, s=100, color='blue', zorder=2)
        ax.text(i, 1.05, layer_names[i], ha='center', va='center', fontsize=12)

    # Draw the connections
    for i in range(len(layers) - 1):
        # if layers[i] == hidden_layers or layers[i+1] == hidden_layers:
            # continue
        y_positions_from = np.linspace(0, 1, layers[i])
        y_positions_to = np.linspace(0, 1, layers[i+1])
        for y1 in y_positions_from:
            for y2 in y_positions_to:
                ax.plot([i, i+1], [y1, y2], color='gray', zorder=1)

    ax.set_xlim(-0.5, len(layers) - 0.5)
    ax.set_ylim(-0.1, 1.1)
    ax.axis('off')
    plt.tight_layout()
    plt.show()

draw_mlp()
```
