```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Define the parameters for the Beta distribution
a, b = 1, 1

# Generate x values
x = np.linspace(0, 1, 1000)

# Compute the PDF for the Beta distribution
y = beta.pdf(x, a, b)

# Plot the Beta(1,1) distribution
plt.plot(x, y, label='Beta')
# plt.title(f'Beta({a},{b}) Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.savefig('beta.eps')
plt.show()
```
